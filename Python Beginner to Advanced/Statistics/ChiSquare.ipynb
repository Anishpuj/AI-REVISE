# Chi-Square Test for Independence

This notebook demonstrates how to perform a Chi-square test of independence to determine if there's a statistically significant relationship between two categorical variables.

## Import Required Libraries

We need numpy for numerical operations, seaborn for dataset loading, scipy.stats for chi-square test functions, and pandas for data manipulation and contingency table creation.

```python
# Import required libraries for statistical analysis
import numpy as np  # For numerical operations and array manipulation
import seaborn as sns  # For loading built-in datasets
from scipy.stats import chi2_contingency  # For chi-square test functions
import pandas as pd  # For data manipulation and contingency table creation
```

## Load Dataset

We'll use the famous Titanic dataset from seaborn to analyze the relationship between gender and survival. This dataset contains information about passengers on the Titanic, including their demographics and survival status.

**Variables of Interest:**
- **sex**: Gender of the passenger (male/female)
- **survived**: Survival status (0 = died, 1 = survived)

This represents a classic case where we want to test if gender was associated with survival chances.

```python
# Load the Titanic dataset from seaborn's built-in datasets
df = sns.load_dataset('titanic')  # Contains passenger demographics and survival data
```

## Explore the Dataset

Let's examine the structure of our dataset to understand the variables we're working with. The Titanic dataset contains 891 passengers with information about their demographics, ticket class, fare, and survival status.

```python
# Display the first few rows of the dataset to understand its structure
df  # Shows passenger demographics and survival information
```

## Create Contingency Table

A contingency table (also called a cross-tabulation table) shows the frequency distribution of variables across categories. This is the foundation for the chi-square test.

**Purpose:**
- Organizes categorical data into a matrix format
- Shows observed frequencies for each combination of categories
- Required input for chi-square test calculations

For our analysis, we'll create a table showing survival counts by gender.

```python
# Create contingency table (cross-tabulation) of sex vs survived
contingency_table = pd.crosstab(df['sex'], df['survived'])  # Shows frequency counts for each gender-survival combination
```

## View the Contingency Table

Let's examine our contingency table to see the observed frequencies:
- **Rows**: Gender categories (female, male)
- **Columns**: Survival status (0 = died, 1 = survived)
- **Values**: Count of passengers in each category combination

```python
# Display the contingency table to observe the frequency distribution
contingency_table  # Shows observed counts for each gender-survival combination
```

## Perform Chi-Square Test

Now we perform the chi-square test of independence using scipy's `chi2_contingency` function. This function returns:
- **chi2**: The chi-square test statistic
- **p-value**: Probability of observing such a distribution by chance if variables are independent
- **dof**: Degrees of freedom (calculated as (rows-1) × (columns-1))
- **expected**: Expected frequencies under the null hypothesis of independence

**Null Hypothesis (H0)**: Gender and survival are independent (no relationship)
**Alternative Hypothesis (H1)**: Gender and survival are related (significant relationship)

```python
# Perform chi-square test of independence on the contingency table
chi2, p_value, dof, expected = chi2_contingency(contingency_table)  # Returns test statistic, p-value, degrees of freedom, and expected frequencies
```

## View Expected Frequencies

The expected frequencies table shows what we would expect to see if gender and survival were truly independent (null hypothesis is true). This is calculated based on the marginal totals of the contingency table.

```python
# Display the expected frequencies under null hypothesis of independence
expected  # Shows what counts we'd expect if gender and survival were unrelated
```

## View Chi-Square Statistic

The chi-square statistic measures the difference between observed and expected frequencies. A larger value suggests a greater deviation from what we'd expect under independence, indicating a potential relationship between variables.

```python
# Display the calculated chi-square test statistic
chi2  # Shows the magnitude of difference between observed and expected frequencies
```

## Set Significance Level

The significance level (alpha) represents our threshold for determining statistical significance. A common choice is 0.05, meaning we're willing to accept a 5% chance of making a Type I error (rejecting a true null hypothesis).

```python
# Set significance level for hypothesis testing
alpha = 0.05  # 5% significance level (common threshold for statistical significance)
```

## Make Statistical Decision

Now we compare our p-value to the significance level (alpha = 0.05) to make a decision:

- **If p-value < alpha**: Reject the null hypothesis (strong evidence of relationship between gender and survival)
- **If p-value >= alpha**: Fail to reject the null hypothesis (insufficient evidence of relationship)

**Interpretation:**
- Rejecting H0 suggests gender was significantly associated with survival chances
- Failing to reject H0 suggests no significant association between gender and survival

```python
# Make statistical decision based on p-value comparison with significance level
if p_value < alpha:
    print("we reject the null hypothesis and there is a significant relationship between gender and survivl")
else:
    print("there is no connenction")  # Insufficient evidence of relationship between gender and survival
```

## Interpretation of Results

Based on our chi-square test analysis:
- **Chi-square statistic**: Measures the deviation between observed and expected frequencies
- **P-value**: Probability of observing such a distribution by chance if variables are independent
- **Decision**: Based on comparison with significance level (alpha = 0.05)

## Key Concepts Summary

1. **Chi-Square Test of Independence**: Tests if two categorical variables are related
2. **Contingency Table**: Cross-tabulation showing observed frequencies
3. **Expected Frequencies**: What we'd expect if variables were independent
4. **Degrees of Freedom**: (rows-1) × (columns-1) for independence tests
5. **P-value Interpretation**: Probability under null hypothesis

## When to Use Chi-Square Test:
- Testing relationship between two categorical variables
- Sample size is sufficiently large (expected frequencies ≥ 5)
- Data consists of counts or frequencies
- Variables are nominal or ordinal

## Practical Applications:
- Market research (customer preferences vs demographics)
- Medical research (treatment vs outcomes)
- Social sciences (education level vs income category)
- Quality control (defect types vs production shifts)

## Assumptions:
- Random sampling
- Independent observations
- Mutually exclusive categories
- Sufficient sample size
