# Outlier Detection Using IQR Method

This notebook demonstrates how to detect and remove outliers from a dataset using the Interquartile Range (IQR) method.

## Import Libraries and Create Sample Data

First, we import the numpy library and create a sample array containing mostly normal values with one obvious outlier (201).

```python
import numpy as np 
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
import numpy as np 
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

## Calculate Quartiles

The IQR method requires calculating the first quartile (Q1) and third quartile (Q3) of the dataset:
- **Q1 (25th percentile)**: The value below which 25% of the data falls
- **Q3 (75th percentile)**: The value below which 75% of the data falls

```python
Q1 = np.percentile(arr,25)
Q3 = np.percentile(arr,75)
```

## Calculate Interquartile Range (IQR)

The IQR is the range between the first and third quartiles. It represents the middle 50% of the data and is used to measure statistical dispersion.

```python
IQR = Q3 - Q1
IQR
```

## Determine Outlier Boundaries

Using the IQR method, we calculate the upper and lower fences:
- **Upper Fence (UF)**: Q3 + 1.5 × IQR - any value above this is considered an outlier
- **Lower Fence (LF)**: Q1 - 1.5 × IQR - any value below this is considered an outlier

The multiplier 1.5 is a standard threshold in statistics for identifying mild outliers.

```python
UF = Q3 + (1.5*IQR)
LF = Q1 - (1.5*IQR)
```

## Filter Out Outliers

Now we iterate through the original array and keep only the values that fall within the calculated boundaries (between LF and UF). Values outside this range are identified as outliers and excluded.

```python
l = []
for i in arr:
    if i >= LF and i <= UF:
        l.append(i)
arr2 = np.array(l)
```

## Compare Original vs Cleaned Data

Let's print both arrays to see the difference:
- **Original array**: Contains all values including the outlier (201)
- **Cleaned array**: Contains only values within the normal range, with the outlier removed

```python
print(arr)
print(arr2)
```

## Visualize the Cleaned Data

A boxplot is an excellent way to visualize the distribution of data and identify outliers. The box represents the IQR, the line inside the box is the median, and the whiskers extend to the fences. Any points beyond the whiskers would be outliers (though in our cleaned dataset, none remain).

```python
import seaborn as sns 

sns.boxplot(x = arr2)
```
