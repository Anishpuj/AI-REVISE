# One-Way ANOVA (Analysis of Variance)

This notebook demonstrates how to perform a one-way ANOVA to determine if there's a statistically significant difference between the means of three or more independent groups.

## Import Required Libraries

We need seaborn for dataset loading, pandas for data manipulation, and scipy.stats for the one-way ANOVA test function.

```python
# Import required libraries for statistical analysis
import seaborn as sns  # For loading built-in datasets
import pandas as pd  # For data manipulation and filtering
from scipy.stats import f_oneway  # For one-way ANOVA test function
```

## Load Dataset

We'll use the Titanic dataset from seaborn to analyze age differences across passenger classes. This dataset contains information about passengers, including their age and ticket class.

**Variables of Interest:**
- **age**: Age of the passenger (continuous variable)
- **pclass**: Passenger class (1 = First, 2 = Second, 3 = Third)

This represents a scenario where we want to test if there's a significant difference in average age across different passenger classes.

```python
# Load the Titanic dataset from seaborn's built-in datasets
df = sns.load_dataset('titanic')  # Contains passenger demographics and survival data
```

## Explore the Dataset

Let's examine the structure of our dataset to understand the variables we're working with. The Titanic dataset contains 891 passengers with information about their demographics, ticket class, and survival status.

```python
# Display the first few rows of the dataset to understand its structure
df  # Shows passenger demographics and survival information
```

## Prepare Data for ANOVA

For one-way ANOVA, we need to:
1. Select only the variables of interest (age and pclass)
2. Remove missing values (NaN) from the age column
3. Create separate groups for each passenger class

**Why Remove Missing Values:**
- ANOVA requires complete data for all groups
- Missing values can bias the results and cause errors
- We'll use dropna() to remove rows with missing age data

```python
# Select only age and pclass columns, then remove missing values from age column
df = df[['age', 'pclass']].dropna()  # Filter to relevant variables and remove NaN values
```

## View Unique Passenger Classes

Let's check the unique values in the pclass column to understand our groups. We'll have three groups representing different passenger classes on the Titanic.

```python
# Display unique passenger classes in the dataset
df['pclass'].unique()  # Shows the three groups: 1st, 2nd, and 3rd class
```

## Create Separate Groups for ANOVA

Now we'll create separate arrays for each passenger class containing the ages. This is required for the one-way ANOVA test, which compares means across multiple groups.

**Groups Created:**
- **class_1**: Ages of passengers in First Class
- **class_2**: Ages of passengers in Second Class  
- **class_3**: Ages of passengers in Third Class

Each group will be treated as an independent sample for the ANOVA analysis.

```python
# Create separate age arrays for each passenger class
class_1 = df[df['pclass'] == 1]['age']  # Ages of First Class passengers
class_2 = df[df['pclass'] == 2]['age']  # Ages of Second Class passengers
class_3 = df[df['pclass'] == 3]['age']  # Ages of Third Class passengers
class_1  # Display First Class ages
```

## Perform One-Way ANOVA

Now we perform the one-way ANOVA test using scipy's `f_oneway` function. This function returns:
- **F-statistic**: Measures the ratio of between-group variance to within-group variance
- **p-value**: Probability of observing such differences by chance if all group means are equal

**Null Hypothesis (H0)**: All passenger class means are equal (no difference in age between classes)
**Alternative Hypothesis (H1)**: At least one passenger class mean differs from the others

**ANOVA Assumptions:**
- Independence of observations
- Normal distribution of residuals
- Homogeneity of variances between groups
- Random sampling

```python
# Perform one-way ANOVA test on the three passenger class age groups
f_stats, p_value = f_oneway(class_1, class_2, class_3)  # Returns F-statistic and p-value
```

## View F-Statistic

The F-statistic measures the ratio of between-group variance to within-group variance. A larger F-value suggests greater differences between group means relative to the variation within each group.

```python
# Display the calculated F-statistic
f_stats  # Shows the ratio of between-group to within-group variance
```

## View P-Value

The p-value represents the probability of observing such differences between group means by chance if the null hypothesis is true. A small p-value (typically < 0.05) suggests strong evidence against the null hypothesis.

```python
# Display the calculated p-value
p_value  # Shows the probability of observing such differences by chance
```

## Set Significance Level

The significance level (alpha) represents our threshold for determining statistical significance. A common choice is 0.05, meaning we're willing to accept a 5% chance of making a Type I error (rejecting a true null hypothesis).

```python
# Set significance level for hypothesis testing
alpha = 0.05  # 5% significance level (common threshold for statistical significance)
```

## Make Statistical Decision

Now we compare our p-value to the significance level (alpha = 0.05) to make a decision:

- **If p-value < alpha**: Reject the null hypothesis (strong evidence of age differences between passenger classes)
- **If p-value >= alpha**: Fail to reject the null hypothesis (insufficient evidence of differences)

**Interpretation:**
- Rejecting H0 suggests at least one passenger class has a significantly different average age
- Failing to reject H0 suggests no significant difference in average age across passenger classes

```python
# Make statistical decision based on p-value comparison with significance level
if p_value < alpha:
    print("reject the null hypothesis and there is a signafacant difference in atleat one passenger class")
else:
    print("there is no significant difference ")  # Insufficient evidence of age differences between classes
```

## Interpretation of Results

Based on our one-way ANOVA analysis:
- **F-statistic**: Measures the ratio of between-group to within-group variance
- **P-value**: Probability of observing such differences by chance if all group means are equal
- **Decision**: Based on comparison with significance level (alpha = 0.05)

## Key Concepts Summary

1. **One-Way ANOVA**: Tests if means of three or more groups are equal
2. **F-statistic**: Ratio of between-group variance to within-group variance
3. **P-value Interpretation**: Probability under null hypothesis
4. **Post-hoc Tests**: If ANOVA is significant, follow-up tests identify which groups differ
5. **Assumptions**: Independence, normality, and homogeneity of variances

## When to Use One-Way ANOVA:
- Comparing means across three or more independent groups
- Continuous dependent variable (age in our case)
- Categorical independent variable with multiple levels (passenger class)
- Sample sizes may be unequal
- When population parameters are unknown

## Practical Applications:
- Medical research (treatment effects across multiple groups)
- Education research (teaching methods comparison)
- Business analytics (performance across departments)
- Quality control (product variations analysis)
