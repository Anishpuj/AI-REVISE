# T-Test for Population Mean

This notebook demonstrates how to perform a one-sample t-test to determine if there's a statistically significant difference between a sample mean and a known population mean when the population standard deviation is unknown.

## Import Required Libraries

We need numpy for numerical operations and scipy.stats for t-distribution functions and statistical tests.

```python
# Import required libraries for statistical analysis
import numpy as np  # For numerical operations and array manipulation
from scipy import stats  # For t-distribution functions and statistical tests
```

## Define Sample and Population Parameters

Here we set up our test scenario:
- **Sample**: A collection of height measurements (in cm) from 10 individuals
- **Sample mean**: The average height of our sample group
- **Sample standard deviation**: The variability within our sample (using Bessel's correction)
- **Sample size (n)**: Number of observations in our sample
- **Population mean**: The known average height (170 cm) for the larger population

**Key Difference from Z-test**: We use sample standard deviation instead of population standard deviation, making the t-test appropriate when population parameters are unknown.

```python
# Define our test data and population parameters
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]  # Sample height measurements in cm
sample_mean = np.mean(sample)  # Calculate sample mean
sample_std = np.std(sample, ddof=1)  # Calculate sample standard deviation (Bessel's correction)
n = len(sample)  # Get sample size
mean_population = 170  # Known population average height (cm)
```

## Calculate the T-Statistic

The t-statistic measures how many standard errors our sample mean is away from the population mean. The formula is:

**t = (Sample Mean - Population Mean) / (Sample Standard Deviation / sqrt(Sample Size))**

**Key Differences from Z-test:**
- Uses sample standard deviation instead of population standard deviation
- Follows t-distribution instead of normal distribution
- More conservative for small sample sizes
- Accounts for additional uncertainty when estimating population parameters from sample

```python
# Calculate the t-statistic using the formula: t = (sample_mean - population_mean) / (sample_std / sqrt(n))
t_stats = (sample_mean - mean_population) / (sample_std / np.sqrt(n))
t_stats  # Display the calculated t-statistic
```

## Calculate the P-Value

The p-value represents the probability of observing a test statistic as extreme as (or more extreme than) our calculated t-statistic, assuming the null hypothesis is true.

For a two-tailed test, we:
1. Take the absolute value of the t-statistic
2. Find the cumulative probability using the t-distribution with (n-1) degrees of freedom
3. Subtract from 1 to get the upper tail probability
4. Multiply by 2 to account for both tails

**Degrees of Freedom**: df = n - 1, representing the number of independent values in the sample.

```python
# Calculate p-value for two-tailed test: p = 2 * (1 - CDF(|t-statistic|, df=n-1))
p_value = 2 * (1 - stats.t.cdf(abs(t_stats), df=n-1))  # Use t-distribution with n-1 degrees of freedom
alpha = 0.05  # Set significance level (5% chance of Type I error)
```

## Make Statistical Decision

Now we compare our p-value to the significance level (alpha = 0.05) to make a decision:

- **If p-value < alpha**: Reject the null hypothesis (strong evidence against H0)
- **If p-value >= alpha**: Fail to reject the null hypothesis (insufficient evidence)

**Null Hypothesis (H0)**: The sample mean equals the population mean (170 cm)
**Alternative Hypothesis (H1)**: The sample mean differs from the population mean

```python
# Make statistical decision based on p-value comparison with significance level
if p_value < alpha:
    print("reject the null hypothesis")  # Strong evidence against H0
else:
    print("accept the null hypothesis")  # Insufficient evidence to reject H0
```

## Interpretation of Results

Based on our t-test analysis:
- **t-statistic**: Measures the standardized difference between sample and population means
- **p-value**: Probability of observing such a difference by chance
- **Decision**: Based on comparison with significance level (alpha = 0.05)

## Key Concepts Summary

1. **T-test vs Z-test**: Use t-test when population standard deviation is unknown
2. **Degrees of Freedom**: df = n - 1, accounts for sample size limitations
3. **T-distribution**: More conservative than normal distribution for small samples
4. **Bessel's Correction**: ddof=1 provides unbiased estimate of population variance
5. **Two-tailed Test**: Tests for differences in either direction from population mean

## When to Use T-Test:
- Population standard deviation is unknown
- Sample size is small (typically n < 30)
- Data is approximately normally distributed
- Samples are independent and randomly selected
