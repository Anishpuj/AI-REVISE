# Two-Sample T-Test for Independent Samples

This notebook demonstrates how to perform a two-sample t-test (also called independent samples t-test) to determine if there's a statistically significant difference between the means of two independent groups.

## Import Required Libraries

We need numpy for numerical operations and scipy.stats for t-distribution functions and statistical tests.

```python
# Import required libraries for statistical analysis
import numpy as np  # For numerical operations and array manipulation
from scipy import stats  # For t-distribution functions and statistical tests
```

## Define Sample Groups

Here we create two independent groups for comparison:
- **Group A**: Test scores from students using Method A (potentially higher scores)
- **Group B**: Test scores from students using Method B (control group)

This represents a scenario where we want to test if there's a significant difference in performance between two teaching methods or treatments.

**Key Assumptions for Two-Sample T-Test:**
- Samples are independent (no relationship between observations in different groups)
- Data is approximately normally distributed
- Homogeneity of variances (unless using Welch's t-test with equal_var=False)

```python
# Define two independent sample groups for comparison
group_A = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88]  # Test scores from Group A (Method A)
group_B = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83]  # Test scores from Group B (Method B)
```

## Perform Two-Sample T-Test

We use scipy's `ttest_ind` function to perform an independent samples t-test. The function returns:
- **t-statistic**: Measures the difference between group means relative to the variation in the data
- **p-value**: Probability of observing such a difference by chance if the null hypothesis is true

**Key Parameters:**
- `equal_var=False`: Uses Welch's t-test, which doesn't assume equal variances between groups (more robust)
- If `equal_var=True`: Assumes equal variances (Student's t-test)

**Welch's T-Test Advantages:**
- More reliable when group variances are unequal
- Performs well even when variances are equal
- Recommended for most practical applications

```python
# Perform two-sample t-test using Welch's method (equal_var=False)
t_stats, p_value = stats.ttest_ind(group_A, group_B, equal_var=False)  # Returns t-statistic and p-value
```

## Display the T-Statistic

The t-statistic value indicates how large the difference between group means is relative to the variability in the data. A larger absolute t-value suggests a more significant difference between groups.

```python
# Display the calculated t-statistic
t_stats  # Shows the standardized difference between group means
```

## Set Significance Level

The significance level (alpha) represents the threshold for determining statistical significance. A common choice is 0.05, meaning we're willing to accept a 5% chance of making a Type I error (rejecting a true null hypothesis).

```python
# Set significance level for hypothesis testing
alpha = 0.05  # 5% significance level (common threshold for statistical significance)
```

## Make Statistical Decision

Now we compare our p-value to the significance level (alpha = 0.05) to make a decision:

- **If p-value < alpha**: Reject the null hypothesis (strong evidence that group means differ)
- **If p-value >= alpha**: Fail to reject the null hypothesis (insufficient evidence of difference)

**Null Hypothesis (H0)**: The means of Group A and Group B are equal
**Alternative Hypothesis (H1)**: The means of Group A and Group B are different

```python
# Make statistical decision based on p-value comparison with significance level
if p_value < alpha:
    print("I will reject the null hypothesis")  # Strong evidence that group means differ
else:
    print("i will accept the null hypothesis")  # Insufficient evidence to claim difference
```

## Interpretation of Results

Based on our two-sample t-test analysis:
- **t-statistic**: Measures the standardized difference between group means
- **p-value**: Probability of observing such a difference by chance if groups have equal means
- **Decision**: Based on comparison with significance level (alpha = 0.05)

## Key Concepts Summary

1. **Two-Sample T-Test**: Compares means between two independent groups
2. **Independent Samples**: Observations in one group don't affect observations in the other
3. **Welch's T-Test**: More robust version that doesn't assume equal variances
4. **Degrees of Freedom**: Calculated differently for unequal variances (Welch-Satterthwaite equation)
5. **Two-Tailed Test**: Tests for differences in either direction (A > B or B > A)

## When to Use Two-Sample T-Test:
- Comparing means between two independent groups
- Testing effectiveness of different treatments/interventions
- Comparing performance between two populations
- When population standard deviations are unknown
- Sample sizes may be unequal

## Practical Applications:
- A/B testing in marketing
- Clinical trials (treatment vs control groups)
- Educational research (teaching method comparisons)
- Quality control (process comparisons)
