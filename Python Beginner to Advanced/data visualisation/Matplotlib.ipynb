```python
# Import required libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

# Matplotlib Basics

```python
# Create sample data for plotting
x = np.linspace(0,5,11)
x
y = x **3 
y
```

```python
# Basic line plot with title and labels
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("Chacha")
plt.ylabel("Chachi")
```

# Subplots in Matplotlib

```python
# Create multiple subplots in a 2x2 grid
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

# Object Oriented way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches
dpi is the dots-per-inch (pixel per inch).

```python
# Create figure with custom size and two overlapping axes
fig = plt.figure(figsize=(20,8),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

# Types of Plots

```python
# Create scatter plot to show individual data points
plt.scatter(x,y)
```

```python
# Generate random data and create histogram
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
# Create box plot to show data distribution
plt.boxplot(x)
```

# Saving Plots

```python
# Create and save a basic plot to file
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```

# Working with Images

```python
# Import image handling library
import matplotlib.image as mpimg
```

```python
# Load and display an image (requires images.jpeg file)
img = mpimg.imread('images.jpeg')

plt.imshow(img)
plt.axis('off')
```

```python
# Crop and display a portion of the image
cropped_img = img[50:200, 100:300]  # crop: rows 50-200, cols 100-300
plt.imshow(cropped_img)
plt.axis('off')
```
