```python
# Import required libraries for data visualization
import seaborn as sns  # Import seaborn library for statistical data visualization
import matplotlib.pyplot as plt  # Import matplotlib.pyplot for creating plots and charts
```

# Introduction to Distribution Plots

Distribution plots help visualize how quantitative data is spread out.

Common types:
- **distplot/histplot**: Shows frequency distribution
- **jointplot**: Displays relationship between two variables
- **pairplot**: Creates scatter plots for all variable pairs
- **rugplot**: Shows individual data points on an axis

```python
# Load the tips dataset from seaborn
df = sns.load_dataset('tips')  # Load the built-in 'tips' dataset from seaborn library
df  # Display the DataFrame to show the loaded data
```

```python
# Check unique values in the 'size' column
df['size'].unique()  # Get unique values from the 'size' column to see different party sizes
```

```python
# Create side-by-side histograms for total_bill and tip
plt.subplot(1,2,1)  # Create a subplot with 1 row and 2 columns, select the first subplot
sns.histplot(df['total_bill'],bins = 20,kde = True)  # Create histogram of total_bill with 20 bins and KDE line
plt.subplot(1,2,2)  # Select the second subplot in the 1x2 grid
sns.histplot(df['tip'],bins = 20,kde = True)  # Create histogram of tip with 20 bins and KDE line
```

```python
# Create joint plot with scatter plot showing relationship between total_bill and tip
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')  # Create joint plot with scatter plot to show relationship between total_bill and tip
```

```python
# Create joint plot with regression line showing trend between total_bill and tip
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'hex')  # Create joint plot with regression line to show trend between total_bill and tip
```

```python
# Create joint plot with regression line showing trend between total_bill and tip
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')  # Create joint plot with regression line to show trend between total_bill and tip
```

```python
# Create pair plot colored by party size to show relationships between all variables
sns.pairplot(df,hue='sex',palette='rainbow')  # Create pair plot colored by party size to show relationships between all variables
```

```python
# Create rug plot showing distribution of tip amounts as individual marks
sns.rugplot(df['tip'])  # Create rug plot showing distribution of tip amounts as individual marks
```
