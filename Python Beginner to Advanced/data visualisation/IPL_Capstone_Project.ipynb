### Advanced Analytics & Enhanced Insights

#### 1. Time-based Analysis - Performance Trends by Month

```python
# Convert date to datetime and extract month (handling mixed formats)
import re

# Function to standardize date format
def standardize_date(date_str):
    # Add space after month name if missing (e.g., "April11,2022" -> "April 11,2022")
    return re.sub(r'([A-Za-z]+)(\d+)', r'\1 \2', date_str)

# Apply standardization
df['date_cleaned'] = df['date'].apply(standardize_date)

# Convert to datetime with consistent format
df['date'] = pd.to_datetime(df['date_cleaned'], format='%B %d,%Y')
df['month'] = df['date'].dt.strftime('%B')

# Monthly match outcomes
monthly_wins = df.groupby(['month', 'match_winner']).size().unstack(fill_value=0)
print("Monthly Match Wins by Team:")
print(monthly_wins)

# Visualize monthly performance
plt.figure(figsize=(12, 6))
monthly_matches = df['month'].value_counts().sort_index()
sns.lineplot(x=monthly_matches.index, y=monthly_matches.values, marker='o', linewidth=2)
plt.title('Number of Matches Played by Month')
plt.ylabel('Number of Matches')
plt.grid(True, alpha=0.3)
plt.show()
```

#### 2. Team Head-to-Head Analysis Matrix

```python
# Create head-to-head matrix
teams = sorted(set(df['team1'].unique()) | set(df['team2'].unique()))
h2h_matrix = pd.DataFrame(0, index=teams, columns=teams)

for _, match in df.iterrows():
    team1, team2, winner = match['team1'], match['team2'], match['match_winner']
    if winner == team1:
        h2h_matrix.loc[team1, team2] += 1
    elif winner == team2:
        h2h_matrix.loc[team2, team1] += 1

# Remove diagonal (same team vs same team)
np.fill_diagonal(h2h_matrix.values, 0)

print("Head-to-Head Win Matrix (Rows: Winning Team, Columns: Losing Team):")
print(h2h_matrix)

# Heatmap of head-to-head results
plt.figure(figsize=(10, 8))
sns.heatmap(h2h_matrix, annot=True, fmt='d', cmap='RdYlBu_r', 
            square=True, cbar_kws={'label': 'Wins'})
plt.title('Team Head-to-Head Win Matrix')
plt.xlabel('Losing Team')
plt.ylabel('Winning Team')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### 3. Venue-Specific Performance Insights

```python
# Venue performance analysis
venue_stats = df.groupby('venue').agg({
    'first_ings_score': ['mean', 'max'],
    'second_ings_score': ['mean', 'max'],
    'match_id': 'count'
}).round(2)

venue_stats.columns = ['Avg_1st_Innings_Score', 'Max_1st_Innings_Score', 
                       'Avg_2nd_Innings_Score', 'Max_2nd_Innings_Score', 'Matches_Played']
venue_stats = venue_stats.sort_values('Matches_Played', ascending=False)

print("Venue Performance Statistics:")
print(venue_stats)

# Team performance at different venues
team_venue_performance = pd.crosstab(df['venue'], df['match_winner'], margins=True)
print("\nTeam Wins by Venue:")
print(team_venue_performance)

# Visualize average scores by venue
plt.figure(figsize=(14, 8))
venue_avg_scores = venue_stats[['Avg_1st_Innings_Score', 'Avg_2nd_Innings_Score']].head(10)
venue_avg_scores.plot(kind='bar', figsize=(14, 6))
plt.title('Average Innings Scores by Venue')
plt.xlabel('Venue')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend(['1st Innings', '2nd Innings'])
plt.tight_layout()
plt.show()
```

#### 4. Correlation Analysis - Key Relationships

```python
# Create numerical columns for correlation analysis
df['toss_win_match_win'] = (df['toss_winner'] == df['match_winner']).astype(int)
df['won_by_numeric'] = df['won_by'].map({'Runs': 1, 'Wickets': 0})
df['toss_decision_numeric'] = df['toss_decision'].map({'Field': 1, 'Bat': 0})

# Correlation matrix
numeric_cols = ['first_ings_score', 'first_ings_wkts', 'second_ings_score', 
                'second_ings_wkts', 'highscore', 'highest_wickets', 'margin',
                'toss_win_match_win', 'won_by_numeric', 'toss_decision_numeric']

correlation_matrix = df[numeric_cols].corr()

print("Correlation Matrix of Key Variables:")
print(correlation_matrix.round(3))

# Heatmap of correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f', cbar_kws={'label': 'Correlation'})
plt.title('Correlation Matrix of IPL Match Variables')
plt.tight_layout()
plt.show()

# Key insights from correlations
print("\nKey Correlation Insights:")
print(f"Toss winning correlates with match winning: {correlation_matrix.loc['toss_win_match_win', 'won_by_numeric']:.3f}")
print(f"First innings score correlates with margin: {correlation_matrix.loc['first_ings_score', 'margin']:.3f}")
print(f"High score correlates with first innings score: {correlation_matrix.loc['highscore', 'first_ings_score']:.3f}")
```

#### 5. Advanced Statistical Insights & Distributions

```python
# Score distributions
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# First innings score distribution
sns.histplot(df['first_ings_score'], kde=True, ax=axes[0,0], color='skyblue')
axes[0,0].set_title('First Innings Score Distribution')
axes[0,0].axvline(df['first_ings_score'].mean(), color='red', linestyle='--', label=f'Mean: {df["first_ings_score"].mean():.1f}')
axes[0,0].legend()

# Second innings score distribution
sns.histplot(df['second_ings_score'], kde=True, ax=axes[0,1], color='lightcoral')
axes[0,1].set_title('Second Innings Score Distribution')
axes[0,1].axvline(df['second_ings_score'].mean(), color='red', linestyle='--', label=f'Mean: {df["second_ings_score"].mean():.1f}')
axes[0,1].legend()

# Win margin distribution
sns.histplot(df[df['won_by'] == 'Runs']['margin'], kde=True, ax=axes[1,0], color='gold')
axes[1,0].set_title('Win Margin by Runs Distribution')
axes[1,0].axvline(df[df['won_by'] == 'Runs']['margin'].mean(), color='red', linestyle='--', 
                  label=f'Mean: {df[df["won_by"] == "Runs"]["margin"].mean():.1f}')
axes[1,0].legend()

# Wickets lost distribution
sns.histplot(df['first_ings_wkts'], kde=True, ax=axes[1,1], color='lightgreen', alpha=0.7, label='1st Innings')
sns.histplot(df['second_ings_wkts'], kde=True, ax=axes[1,1], color='orange', alpha=0.7, label='2nd Innings')
axes[1,1].set_title('Wickets Lost Distribution')
axes[1,1].legend()

plt.tight_layout()
plt.show()

# Statistical summary
print("Statistical Summary of Key Metrics:")
stats_summary = df[['first_ings_score', 'second_ings_score', 'highscore', 'margin', 'highest_wickets']].describe().round(2)
print(stats_summary)
```

#### 6. Team Performance Deep Dive

```python
# Comprehensive team performance analysis
team_stats = df.groupby('match_winner').agg({
    'match_id': 'count',
    'first_ings_score': lambda x: x[df.loc[x.index, 'won_by'] == 'Runs'].mean() if any(df.loc[x.index, 'won_by'] == 'Runs') else 0,
    'second_ings_score': lambda x: x[df.loc[x.index, 'won_by'] == 'Wickets'].mean() if any(df.loc[x.index, 'won_by'] == 'Wickets') else 0,
    'margin': 'mean',
    'highscore': 'max'
}).round(2)

team_stats.columns = ['Total_Wins', 'Avg_Batting_First_Score', 'Avg_Chasing_Score', 'Avg_Win_Margin', 'Highest_Score']
team_stats = team_stats.sort_values('Total_Wins', ascending=False)

print("Comprehensive Team Performance Statistics:")
print(team_stats)

# Win rate analysis (teams that played vs won)
teams_played = pd.concat([df['team1'], df['team2']]).value_counts()
team_win_rate = (df['match_winner'].value_counts() / teams_played * 100).round(2)
team_win_rate = team_win_rate.sort_values(ascending=False)

print("\nTeam Win Rates (%):")
print(team_win_rate)

# Visualize team performance
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Total wins
team_stats['Total_Wins'].sort_values().plot(kind='barh', ax=axes[0], color='steelblue')
axes[0].set_title('Total Wins by Team')
axes[0].set_xlabel('Wins')

# Win rates
team_win_rate.sort_values().plot(kind='barh', ax=axes[1], color='darkgreen')
axes[1].set_title('Win Rate by Team (%)')
axes[1].set_xlabel('Win Rate (%)')

# Average win margin
team_stats['Avg_Win_Margin'].sort_values().plot(kind='barh', ax=axes[2], color='darkorange')
axes[2].set_title('Average Win Margin by Team')
axes[2].set_xlabel('Average Margin')

plt.tight_layout()
plt.show()
```

#### 8. Tournament Progression Analysis

```python
# Analyze performance progression through the tournament
df['match_day'] = range(1, len(df) + 1)

# Rolling average of scores
df['rolling_avg_1st'] = df['first_ings_score'].rolling(window=10, min_periods=1).mean()
df['rolling_avg_2nd'] = df['second_ings_score'].rolling(window=10, min_periods=1).mean()

# Plot tournament progression
plt.figure(figsize=(14, 8))
plt.plot(df['match_day'], df['first_ings_score'], 'o-', alpha=0.3, label='1st Innings Score', color='blue')
plt.plot(df['match_day'], df['second_ings_score'], 'o-', alpha=0.3, label='2nd Innings Score', color='red')
plt.plot(df['match_day'], df['rolling_avg_1st'], '-', linewidth=3, label='Rolling Avg (1st)', color='darkblue')
plt.plot(df['match_day'], df['rolling_avg_2nd'], '-', linewidth=3, label='Rolling Avg (2nd)', color='darkred')

plt.title('IPL 2022 Tournament Score Progression')
plt.xlabel('Match Number')
plt.ylabel('Score')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Stage-wise performance
stage_performance = df.groupby('stage').agg({
    'first_ings_score': ['mean', 'max'],
    'second_ings_score': ['mean', 'max'],
    'margin': 'mean',
    'match_id': 'count'
}).round(2)

stage_performance.columns = ['Avg_1st_Score', 'Max_1st_Score', 'Avg_2nd_Score', 
                           'Max_2nd_Score', 'Avg_Margin', 'Matches']
print("Stage-wise Performance Analysis:")
print(stage_performance)
```
