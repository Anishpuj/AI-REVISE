# Introduction to Regression Plots (lmplot)

"""
Regression plots are essential tools in data visualization and statistical analysis. 
They help us understand the relationship between two numerical variables by:

1. **Scatter Plot**: Shows individual data points as dots
2. **Regression Line**: A line of best fit that shows the trend in the data
3. **Confidence Interval**: The shaded area around the regression line that indicates uncertainty

## Why Use Regression Plots?
- Identify trends and patterns in data
- Understand correlation between variables
- Make predictions based on the relationship
- Detect outliers and unusual data points

## lmplot() in Seaborn
The `lmplot()` function in seaborn creates a regression plot with:
- Scatter plot of data points
- Linear regression line
- 95% confidence interval by default

While regression analysis is a fundamental concept in machine learning for predictive modeling,
in this tutorial, we'll focus on the visualization aspect and how to create effective regression plots.

## Common Parameters:
- `x`: Variable for x-axis (independent variable)
- `y`: Variable for y-axis (dependent variable)  
- `data`: DataFrame containing the data
- `hue`: Grouping variable that creates separate plots with different colors
- `palette`: Color scheme for the plot
- `markers`: Shape of markers for different groups
- `scatter_kws`: Dictionary of keyword arguments for scatter plot styling
"""

````python
# Load the tips dataset from seaborn
tips = sns.load_dataset("tips")

"""
## About the Tips Dataset:
The 'tips' dataset is a built-in seaborn dataset that contains information about tips received by a waiter
over a period of a few months working in a restaurant. It's perfect for demonstrating regression analysis.

### Dataset Columns:
- **total_bill**: Total bill amount (in dollars)
- **tip**: Tip amount (in dollars) 
- **sex**: Gender of the person paying the bill (Male/Female)
- **smoker**: Whether the party included smokers (Yes/No)
- **day**: Day of the week (Thur, Fri, Sat, Sun)
- **time**: Time of day (Lunch/Dinner)
- **size**: Size of the party (number of people)

### Why This Dataset for Regression?
- Contains numerical variables (total_bill, tip) perfect for regression analysis
- Has categorical variables for grouping and comparison
- Real-world data that's easy to understand
- Shows realistic relationship between bill amount and tip

Let's explore the relationship between total bill and tip amount!
"""

# Display first few rows to understand the data structure
print("First 5 rows of the tips dataset:")
print(tips.head())
print(f"\nDataset shape: {tips.shape}")
print(f"\nColumn names: {list(tips.columns)}")
````

````python
# Basic Regression Plot
sns.lmplot(x='total_bill', y='tip', data=tips)

"""
## Basic Regression Plot Analysis

This plot shows the relationship between total bill amount and tip amount:

### What We Can Observe:
1. **Positive Correlation**: As the total bill increases, the tip amount tends to increase
2. **Regression Line**: The blue line shows the average relationship - for every dollar increase in bill, tip increases by about $0.10-0.15
3. **Confidence Interval**: The light blue shaded area represents the 95% confidence interval
   - Wider intervals indicate more uncertainty in predictions
   - Narrower intervals indicate more confident predictions

### Key Insights:
- Most data points cluster around the regression line
- Some outliers exist (people who tipped much more or less than expected)
- The relationship appears linear but not perfectly correlated
- There's variability in tipping behavior even for similar bill amounts

### Interpretation:
- The upward slope confirms that higher bills generally lead to higher tips
- The spread of points shows that other factors also influence tipping behavior
- This is a typical pattern in service industry tipping

The plot automatically calculates and displays the linear regression line using ordinary least squares (OLS) method.
"""

plt.title('Relationship Between Total Bill and Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# Regression Plot with Gender Differentiation
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')

"""
## Regression Plot by Gender

This plot separates the data by gender (male/female) to see if tipping behavior differs:

### Key Observations:
1. **Separate Regression Lines**: 
   - Blue line: Male customers
   - Orange line: Female customers
2. **Different Patterns**: Each gender may have different tipping behaviors
3. **Color Coding**: Makes it easy to distinguish between groups

### Analysis by Gender:
- **Male Customers** (Blue): May show different slope and intercept
- **Female Customers** (Orange): May have different tipping patterns
- **Overlap**: Some behaviors are similar across genders

### Why Use 'hue' Parameter:
- Enables comparison between different groups
- Helps identify if relationships differ by category
- Useful for discovering patterns that might be hidden in aggregated data
- Essential for understanding demographic differences in behavior

### Business Insights:
- If tipping differs by gender, restaurants might adjust service strategies
- Helps in understanding customer segmentation
- Can inform marketing and service improvements

The 'hue' parameter is powerful for exploring categorical variables and their impact on relationships.
"""

plt.title('Tipping Behavior by Gender')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.grid(True, alpha=0.3)
plt.legend(title='Gender')
plt.show()
````

````python
# Regression Plot with Custom Color Palette
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')

"""
## Custom Color Palette

This plot uses the 'coolwarm' color palette to enhance visual appeal and readability:

### Color Palette Options in Seaborn:
- **'coolwarm'**: Blue to red gradient (good for showing contrast)
- **'viridis', 'plasma', 'inferno', 'magma'**: Perceptually uniform color maps
- **'Set1', 'Set2', 'Set3'**: Categorical color palettes
- **'deep', 'muted', 'pastel', 'bright', 'dark'**: Various color schemes
- Custom colors: ['#FF6B6B', '#4ECDC4'] (hex codes)

### Why Color Choice Matters:
1. **Accessibility**: Choose colors that are distinguishable by colorblind users
2. **Professional Appearance**: Appropriate colors for presentations/reports
3. **Emotional Impact**: Colors can convey different meanings
4. **Print Compatibility**: Some colors work better in print vs digital

### 'coolwarm' Palette Benefits:
- High contrast between categories
- Professional and modern appearance
- Works well for binary classifications
- Good for both screen and print

### Best Practices:
- Consider your audience and context
- Ensure colors are distinguishable
- Test with colorblind simulators
- Be consistent across related visualizations

Color customization helps create more professional and publication-ready plots.
"""

plt.title('Tipping Behavior by Gender (Coolwarm Palette)')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.grid(True, alpha=0.3)
plt.legend(title='Gender')
plt.show()
````

````python
# Advanced Regression Plot with Custom Markers and Styling
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', 
           markers=['o', 'v'], scatter_kws={'s': 25})

"""
## Advanced Customization Options

This plot demonstrates additional customization options for professional visualizations:

### Custom Markers:
- **'o'**: Circle markers (for Male customers)
- **'v'**: Triangle markers (for Female customers)
- Other marker options: 's' (square), '^' (triangle up), 'D' (diamond), '*' (star), etc.

### scatter_kws Parameter:
- **'s': 50**: Sets marker size to 50 points
- Other scatter_kws options:
  - 'alpha': transparency (0.0 to 1.0)
  - 'edgecolor': color of marker edges
  - 'linewidth': width of marker edges
  - 'facecolor': fill color of markers

### Additional Customization Options:
```python
# Line customization
line_kws = {
    'linewidth': 2,      # Line thickness
    'linestyle': '--',   # Line style
    'alpha': 0.8         # Line transparency
}

# Plot size and aspect
sns.lmplot(..., height=6, aspect=1.5)

# Grid and background
sns.set_style("whitegrid")  # or "darkgrid", "white", "dark", "ticks"
```

### Professional Tips:
1. **Consistency**: Use consistent styling across related plots
2. **Clarity**: Ensure markers are distinguishable when overlapped
3. **Size**: Adjust marker size based on data density
4. **Color**: Choose colors that work in both color and grayscale

### When to Use Advanced Customization:
- Academic papers and publications
- Business presentations
- Reports and documentation
- When creating a consistent visual identity

These customization options help create publication-quality visualizations that effectively communicate your data insights.
"""

plt.title('Advanced Tipping Analysis by Gender')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.grid(True, alpha=0.3)
plt.legend(title='Gender', labels=['Male', 'Female'])
plt.tight_layout()
plt.show()

# Additional analysis: Correlation coefficients
male_data = tips[tips['sex'] == 'Male']
female_data = tips[tips['sex'] == 'Female']

print("Correlation Analysis:")
print(f"Male customers: {male_data['total_bill'].corr(male_data['tip']):.3f}")
print(f"Female customers: {female_data['total_bill'].corr(female_data['tip']):.3f}")
print(f"Overall: {tips['total_bill'].corr(tips['tip']):.3f}")
````
