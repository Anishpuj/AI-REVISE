```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df['Months'] = df['Total Time'].apply(calculate_total_months)
```

# NEW FEATURE EXTRACTION

```python
# Advanced Feature Extraction Techniques

# Let's extract more meaningful features from the anime dataset
# We'll extract: Type, Members, Year, Season, Duration, and more

# Extract anime type (TV, Movie, OVA, etc.)
def extract_anime_type(title):
    """Extract the type of anime from title"""
    if 'TV' in title:
        return 'TV'
    elif 'Movie' in title:
        return 'Movie'
    elif 'OVA' in title:
        return 'OVA'
    elif 'Special' in title:
        return 'Special'
    else:
        return 'Unknown'

df['Type'] = df['Title'].apply(extract_anime_type)
print("Anime types extracted:")
print(df['Type'].value_counts())
```

```python
# Extract member count (number of people who watched)
import re

def extract_members(title):
    """Extract member count from title"""
    # Look for patterns like "1,034,410 members" or "605,113 members"
    match = re.search(r'(\d{1,3}(,\d{3})*\s*members)', title)
    if match:
        # Remove commas and 'members' text, convert to int
        members_str = match.group(1).replace(',', '').replace('members', '').strip()
        return int(members_str)
    return 0

df['Members'] = df['Title'].apply(extract_members)
print(f"Average members: {df['Members'].mean():,.0f}")
print(f"Max members: {df['Members'].max():,.0f}")
print(f"Min members: {df['Members'].min():,.0f}")
```

```python
# Extract start year and create decade feature
def extract_start_year(title):
    """Extract the start year from title"""
    # Look for 4-digit year patterns
    match = re.search(r'\b(19|20)\d{2}\b', title)
    if match:
        return int(match.group())
    return 0

df['Start_Year'] = df['Title'].apply(extract_start_year)

# Create decade feature
df['Decade'] = (df['Start_Year'] // 10) * 10
df['Decade'] = df['Decade'].astype(str) + 's'

print("Anime by decade:")
print(df['Decade'].value_counts().sort_index())
```

```python
# Extract season information
def extract_season(title):
    """Extract season from title"""
    seasons = ['Winter', 'Spring', 'Summer', 'Fall', 'Autumn']
    for season in seasons:
        if season in title:
            return season
    return 'Unknown'

df['Season'] = df['Title'].apply(extract_season)
print("Anime by season:")
print(df['Season'].value_counts())
```

```python
# Create derived features for better analysis

# Episodes per month (frequency)
df['Episodes_Per_Month'] = df.apply(lambda row: row['Episodes'] / row['Months'] if row['Months'] > 0 else 0, axis=1)

# Score categories
def categorize_score(score):
    """Categorize anime by score"""
    if score >= 9.0:
        return 'Excellent'
    elif score >= 8.5:
        return 'Very Good'
    elif score >= 8.0:
        return 'Good'
    elif score >= 7.5:
        return 'Average'
    else:
        return 'Below Average'

df['Score_Category'] = df['Score'].apply(categorize_score)

# Length categories
def categorize_length(episodes):
    """Categorize anime by length"""
    if episodes == 1:
        return 'Movie'
    elif episodes <= 12:
        return 'Short Series'
    elif episodes <= 26:
        return 'Standard Series'
    elif episodes <= 52:
        return 'Long Series'
    else:
        return 'Very Long Series'

df['Length_Category'] = df['Episodes'].apply(categorize_length)

print("Feature categories created:")
print("Score Categories:", df['Score_Category'].value_counts())
print("\nLength Categories:", df['Length_Category'].value_counts())
```

```python
# Create popularity score (normalized members)
df['Popularity_Score'] = (df['Members'] - df['Members'].min()) / (df['Members'].max() - df['Members'].min())

# Create engagement score (combination of score and popularity)
df['Engagement_Score'] = (df['Score'] / 10) * 0.7 + df['Popularity_Score'] * 0.3

# Extract franchise information (detect sequels/prequels)
def detect_franchise(title):
    """Detect if anime is part of a franchise"""
    franchise_keywords = ['Season', 'Part', '2nd', '3rd', '4th', 'Final', 'The Final', 'Movie', 'II', 'III']
    for keyword in franchise_keywords:
        if keyword in title:
            return True
    return False

df['Is_Franchise'] = df['Title'].apply(detect_franchise)

print("Derived features created:")
print(f"Franchise anime: {df['Is_Franchise'].sum()} out of {len(df)}")
print(f"Average engagement score: {df['Engagement_Score'].mean():.3f}")
print(f"Average popularity score: {df['Popularity_Score'].mean():.3f}")
```
