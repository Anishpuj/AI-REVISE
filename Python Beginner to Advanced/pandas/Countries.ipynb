```python
# Advanced Data Extraction and Feature Engineering

# Let's extract more meaningful features from the countries dataset
# We'll create economic indicators, development metrics, and geographical features
```

```python
# Create economic indicators and development metrics

# GDP per capita
df['gdp_per_capita'] = df['gdp'] / df['population']

# Population density (people per square km)
df['population_density'] = df['population'] / df['land_area']

# Urbanization rate
df['urbanization_rate'] = (df['urban_population'] / df['population']) * 100

# Gender balance ratio (female to male ratio)
df['gender_ratio'] = df['population_female'] / df['population_male']

# Dependency ratio (assuming working age 15-64, but using median age as proxy)
df['dependency_ratio_proxy'] = (df['population'] - (df['median_age'] * df['population'] / 25)) / (df['median_age'] * df['population'] / 25)

print("Economic and Demographic Indicators Created:")
print(f"GDP per capita range: ${df['gdp_per_capita'].min():,.0f} - ${df['gdp_per_capita'].max():,.0f}")
print(f"Population density range: {df['population_density'].min():.1f} - {df['population_density'].max():.1f} people/km²")
print(f"Urbanization rate range: {df['urbanization_rate'].min():.1f}% - {df['urbanization_rate'].max():.1f}%")
```

```python
# Create health and development indices

# Health Expenditure per capita
df['health_expenditure_per_capita'] = (df['health_expenditure_pct_gdp'] / 100) * df['gdp_per_capita']

# Hospital beds per 1000 people
df['hospital_beds_per_1000'] = (df['hospital_beds'] / df['population']) * 1000

# Internet penetration score (0-100 scale)
df['internet_score'] = df['internet_pct']

# Press freedom score (inverted - higher is better)
df['press_freedom_score'] = 10 - df['press']

# Life expectancy score (normalized 0-100)
df['life_expectancy_score'] = (df['life_expectancy'] / 85) * 100

# Women's representation score
df['women_representation_score'] = df['women_parliament_seats_pct']

print("Health and Development Indices Created:")
print(f"Health expenditure per capita range: ${df['health_expenditure_per_capita'].min():.0f} - ${df['health_expenditure_per_capita'].max():.0f}")
print(f"Hospital beds per 1000 range: {df['hospital_beds_per_1000'].min():.1f} - {df['hospital_beds_per_1000'].max():.1f}")
print(f"Life expectancy score range: {df['life_expectancy_score'].min():.1f} - {df['life_expectancy_score'].max():.1f}")
```

```python
# Create environmental and energy metrics

# CO2 emissions per capita
df['co2_per_capita'] = df['co2_emissions'] / df['population']

# CO2 intensity (CO2 per GDP)
df['co2_intensity'] = df['co2_emissions'] / df['gdp']

# Renewable energy adoption score
df['renewable_energy_score'] = df['renewable_energy_consumption_pct']

# Energy self-sufficiency (inverse of imports)
df['energy_self_sufficiency'] = 100 - df['energy_imports_pct']

# Environmental impact score (combination of emissions and renewable energy)
df['environmental_score'] = (df['renewable_energy_score'] * 0.6) + ((100 - df['co2_per_capita'] / df['co2_per_capita'].max() * 100) * 0.4)

print("Environmental Metrics Created:")
print(f"CO2 per capita range: {df['co2_per_capita'].min():.2f} - {df['co2_per_capita'].max():.2f}")
print(f"CO2 intensity range: {df['co2_intensity'].min():.6f} - {df['co2_intensity'].max():.6f}")
print(f"Environmental score range: {df['environmental_score'].min():.1f} - {df['environmental_score'].max():.1f}")
```

```python
# Create geographical and regional features

# Hemisphere classification
def classify_hemisphere(lat):
    if lat > 0:
        return 'Northern'
    elif lat < 0:
        return 'Southern'
    else:
        return 'Equatorial'

df['hemisphere'] = df['latitude'].apply(classify_hemisphere)

# Climate zone based on latitude
def classify_climate_zone(lat):
    abs_lat = abs(lat)
    if abs_lat < 23.5:
        return 'Tropical'
    elif abs_lat < 40:
        return 'Subtropical'
    elif abs_lat < 60:
        return 'Temperate'
    else:
        return 'Polar'

df['climate_zone'] = df['latitude'].apply(classify_climate_zone)

# Distance from equator
df['distance_from_equator'] = abs(df['latitude'])

# Region size (number of countries per region)
region_counts = df['region'].value_counts()
df['region_size'] = df['region'].map(region_counts)

print("Geographical Features Created:")
print(f"Hemispheres: {df['hemisphere'].value_counts().to_dict()}")
print(f"Climate zones: {df['climate_zone'].value_counts().to_dict()}")
print(f"Region sizes range: {df['region_size'].min()} - {df['region_size'].max()} countries")
```

```python
# Create development and governance indices

# Human Development Index (simplified version)
# Components: life expectancy, education (internet proxy), income (GDP per capita)
df['life_exp_normalized'] = df['life_expectancy'] / 85
df['education_normalized'] = df['internet_pct'] / 100
df['income_normalized'] = np.log(df['gdp_per_capita']) / np.log(df['gdp_per_capita'].max())
df['simplified_hdi'] = (df['life_exp_normalized'] + df['education_normalized'] + df['income_normalized']) / 3

# Development level classification
def classify_development(hdi):
    if hdi >= 0.8:
        return 'Very High'
    elif hdi >= 0.7:
        return 'High'
    elif hdi >= 0.55:
        return 'Medium'
    else:
        return 'Low'

df['development_level'] = df['simplified_hdi'].apply(classify_development)

# Governance score (democracy + press freedom)
df['governance_score'] = (df['democracy_score'] / 10) * 0.7 + (df['press_freedom_score'] / 10) * 0.3

# Economic stability score (inverse of inflation + debt management)
df['economic_stability'] = ((100 - df['inflation']) / 100) * 0.6 + ((100 - df['central_government_debt_pct_gdp']) / 100) * 0.4

print("Development Indices Created:")
print(f"HDI range: {df['simplified_hdi'].min():.3f} - {df['simplified_hdi'].max():.3f}")
print(f"Development levels: {df['development_level'].value_counts().to_dict()}")
print(f"Governance score range: {df['governance_score'].min():.3f} - {df['governance_score'].max():.3f}")
```

```python
# Create composite rankings and scores

# Overall Development Score (weighted combination)
df['overall_development_score'] = (
    df['simplified_hdi'] * 0.3 +
    df['governance_score'] * 0.2 +
    df['economic_stability'] * 0.2 +
    df['environmental_score'] / 100 * 0.15 +
    df['internet_score'] / 100 * 0.15
)

# Quality of Life Index
df['quality_of_life_index'] = (
    df['life_expectancy_score'] / 100 * 0.3 +
    df['internet_score'] / 100 * 0.2 +
    df['press_freedom_score'] / 10 * 0.2 +
    df['women_representation_score'] / 50 * 0.15 +
    df['health_expenditure_per_capita'] / df['health_expenditure_per_capita'].max() * 0.15
)

# Economic Power Index
df['economic_power_index'] = (
    df['gdp'] / df['gdp'].max() * 0.4 +
    df['gdp_per_capita'] / df['gdp_per_capita'].max() * 0.3 +
    (100 - df['unemployment_pct']) / 100 * 0.2 +
    (100 - df['inflation']) / 100 * 0.1
)

print("Composite Indices Created:")
print(f"Overall Development Score range: {df['overall_development_score'].min():.3f} - {df['overall_development_score'].max():.3f}")
print(f"Quality of Life Index range: {df['quality_of_life_index'].min():.3f} - {df['quality_of_life_index'].max():.3f}")
print(f"Economic Power Index range: {df['economic_power_index'].min():.3f} - {df['economic_power_index'].max():.3f}")
```
